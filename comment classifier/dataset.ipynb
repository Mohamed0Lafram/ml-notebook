# predict

```python
def clean_comment(comment):
    #delete special characters
    comment = re.sub(r'[^a-zA-Z0-9\'/ ]', '',comment)

    #delete comlemnetary words
    comment = re.sub(stop_words_pattern, '', comment, flags=re.IGNORECASE)

    return comment
```

```python
def predict(comment):
    #clean the comment
    comment = clean_comment(comment)

    #this function will get the proba of the coment being positive and then the proba of the comment being negative then it will compare them and return the best proba
    comment = comment.split()
    
    #positve proba 
    p_y = train[train['sentiment']== 'positive'].count()[0] / train.count()[0]
    positive_proba_coment = p_y 
    for c in comment:
        if c in p_words_proba:
            positive_proba_coment *= p_words_proba[c]
        else: 
            positive_proba_coment *= 1/p_frequency.sum()
        
    #positve proba 
    n_y = train[train['sentiment']== 'negative'].count()[0] / train.count()[0]
    negative_proba_coment = n_y
    for c in comment:
        if c in n_words_proba:
            negative_proba_coment *= n_words_proba[c]
        else: 
            negative_proba_coment *= 1/n_frequency.sum()

    #print(negative_proba_coment)
    #print(positive_proba_coment)
    
    if negative_proba_coment > positive_proba_coment:
        #print('le commentaire est negative')
        return 0
    else : 
        #print('le commentaire est positive')
        return 1
```

```python
#alterantive predict function with a less cost
import numpy as np

# Precompute these values once instead of inside predict()
p_y = train['sentiment'].value_counts(normalize=True)['positive']
n_y = train['sentiment'].value_counts(normalize=True)['negative']

def predict1(comment):
    # Clean the comment
    comment = clean_comment(comment)
    
    # Tokenize the comment
    words = comment.split()

    # Log probability of positive sentiment
    positive_log_proba = np.log(p_y)  
    for word in words:
        positive_log_proba += np.log(p_words_proba.get(word, 1 / p_frequency.sum()))  # Use `.get()`

    # Log probability of negative sentiment
    negative_log_proba = np.log(n_y)
    for word in words:
        negative_log_proba += np.log(n_words_proba.get(word, 1 / n_frequency.sum()))

    # Compare log probabilities instead of raw multiplication
    return 1 if positive_log_proba > negative_log_proba else 0
```

```python
def accurecy():
    #predict all 
    #if the prediction is correct add one else 0 
    #return the percesnt of the acuresy
    tp,tn,fp,fn = 0,0,0,0
    for i in range(40000,len(test)+40000):
        prediction = predict1(test.loc[i,'review'])
        if prediction == 1 and test.loc[i,'sentiment'] == prediction:
            tp +=1
        elif prediction == 0 and test.loc[i,'sentiment'] == prediction:
            tn +=1
        elif prediction == 1 and test.loc[i,'sentiment'] != prediction:
            fp += 1
        else:
            fn += 1
        percentage = (tp+tn) / (tp+tn+fp+fn)
    return percentage
```
